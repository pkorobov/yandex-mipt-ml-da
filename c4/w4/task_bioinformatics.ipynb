```python
import pandas as pd
import scipy.stats
import numpy as np
```

```python
df = pd.read_csv("gene_high_throughput_sequencing.csv")
```

```python
df['Diagnosis'].value_counts()
```

```python
print(df.shape)
df.head()
```

```python
" ".join(df.columns)
```

```python
X1 = df[df['Diagnosis'] == 'normal'].iloc[:, 2:]
X2 = df[df['Diagnosis'] == 'early neoplasia'].iloc[:, 2:]
X3 = df[df['Diagnosis'] == 'cancer'].iloc[:, 2:]
```

```python
ans = 0
p_values_1 = []
for i in range(X1.shape[1]):
    p_values_1.append(scipy.stats.ttest_ind(X1.iloc[:, i], X2.iloc[:, i], equal_var=False)[1])
    ans += scipy.stats.ttest_ind(X1.iloc[:, i], X2.iloc[:, i], equal_var=False)[1] < 0.05
with open('1.1.txt', 'w') as file:
    file.write(str(ans))
ans
```

```python
ans = 0
p_values_2 = []
for i in range(X1.shape[1]):
    p_values_2.append(scipy.stats.ttest_ind(X2.iloc[:, i], X3.iloc[:, i], equal_var=False)[1])
    ans += scipy.stats.ttest_ind(X2.iloc[:, i], X3.iloc[:, i], equal_var=False)[1] < 0.05
with open('1.2.txt', 'w') as file:
    file.write(str(ans))
ans
```

```python
import statsmodels.stats.multitest as smm
```

```python
mod1 = smm.multipletests(p_values_1, 
                        alpha = 0.025, 
                        method = 'holm')
mod2 = smm.multipletests(p_values_2, 
                        alpha = 0.025, 
                        method = 'holm')

def fold_change(control, treatment):
    C = control.values.mean(axis=0)
    T = treatment.values.mean(axis=0)
    return np.where(T > C, T / C, -C / T)


ans1 = np.sum(np.abs(fold_change(X1, X2)[mod1[0]]) > 1.5)
with open('2.1.txt', 'w') as file:
    file.write(str(ans1))

ans2 = np.sum(np.abs(fold_change(X2, X3)[mod2[0]]) > 1.5)    
with open('2.2.txt', 'w') as file:
    file.write(str(ans2))
    
ans1, ans2
```

```python
mod1 = smm.multipletests(p_values_1, 
                        alpha = 0.025, 
                        method = 'fdr_bh')
mod2 = smm.multipletests(p_values_2, 
                        alpha = 0.025, 
                        method = 'fdr_bh')

def fold_change(control, treatment):
    C = control.values.mean(axis=0)
    T = treatment.values.mean(axis=0)
    return np.where(T > C, T / C, -C / T)


ans1 = np.sum(np.abs(fold_change(X1, X2)[mod1[0]]) > 1.5)
with open('3.1.txt', 'w') as file:
    file.write(str(ans1))

ans2 = np.sum(np.abs(fold_change(X2, X3)[mod2[0]]) > 1.5)    
with open('3.2.txt', 'w') as file:
    file.write(str(ans2))
    
ans1, ans2
```
